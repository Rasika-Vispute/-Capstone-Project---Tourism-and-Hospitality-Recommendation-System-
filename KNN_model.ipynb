```python
df_final = df
```

```python
df_final.set_index('Unnamed: 0',inplace=True)
```

```python
# df_final = df_final.drop('Unnamed: 0',axis=1)
```

```python
df_final = df_final.fillna(0)
df_final
```

```python
df_final = df_final.T
df_final
```

```python
from scipy.sparse import csr_matrix

city_features_df_matrix = csr_matrix(df_final.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine',algorithm='brute')
```

```python
model_knn.fit(city_features_df_matrix)
```

```python
df_final.shape
```

```python
query_index = np.random.choice(df_final.shape[0])
print(query_index)
```

```python
# query_index = 353
```

```python
# df_final
```

```python
distances,indices = model_knn.kneighbors(df_final.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('Recommendations for {0}: \n'.format(df_final.index[query_index]))
    else:
        print('{0}:{1},with distance of {2}'.format(i,df_final.index[indices.flatten()[i]],distances.flatten()[i]))
```
